```python
import speech_recognition as sr
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def capture_audio(interval):
    recognizer = sr.Recognizer()
    microphone = sr.Microphone()

    with microphone as source:
        print("Capture iniciada. Fale algo...")
        while True:
            try:
                audio = recognizer.listen(source, timeout=interval)
                text = recognizer.recognize_google(audio, language='pt-BR')
                print("Texto reconhecido: ", text)
                generate_word_cloud(text)
            except sr.UnknownValueError:
                print("Não foi possível reconhecer o áudio.")
            except sr.RequestError:
                print("Erro na requisição ao serviço de reconhecimento de fala.")
            except sr.WaitTimeoutError:
                print(f"Nenhum áudio detectado nos últimos {interval} segundos.")

def generate_word_cloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Defina o intervalo de captura em segundos
intervalo_captura = 5

# Chame a função para capturar áudio com o intervalo definido
capture_audio(intervalo_captura)
```
